```python
import os
from deepvoice import DeepVoice

# Initialize DeepVoice API
print("Initializing DeepVoice...")
DeepVoice.initialize()
print("Initialization complete.")

# Define file paths
audio_path = os.path.join(os.getcwd(), "deeptest", "audio", "tedtalk.mp3")
voice1_path = os.path.join(os.getcwd(), "deeptest", "audio", "billgates1.mp3")
voice2_path = os.path.join(os.getcwd(), "deeptest", "audio", "billgates1.mp3")

print(f"Audio path: {audio_path}")
print(f"Voice 1 path: {voice1_path}")
print(f"Voice 2 path: {voice2_path}")
```

```python
print("\n=== Testing voice extraction ===")
try:
    print(f"Extracting voices from {audio_path}...")
    voices = DeepVoice.extract_voices(audio_path)
    print(f"Found {len(voices)} voices.")

    if voices:
        print("\nFirst voice properties:")
        voice = voices[0]
        print(f"  - Speaker ID: {voice.speaker_id if hasattr(voice, 'speaker_id') else 'N/A'}")
        print(f"  - Duration: {voice.duration if hasattr(voice, 'duration') else 'N/A'}")
        print(f"  - Confidence: {voice.confidence if hasattr(voice, 'confidence') else 'N/A'}")
        print(f"  - Audio shape: {voice.audio.shape}")
except Exception as e:
    print(f"Error in voice extraction: {e}")
```

```python
print("\n=== Testing voice verification ===")
try:
    print(f"Verifying if {voice1_path} and {voice2_path} are the same voice...")
    result = DeepVoice.verify(voice1_path, voice2_path)

    print("\nVerification result:")
    print(f"  - Is same voice: {result['is_same']}")
    print(f"  - Confidence: {result['confidence']}")
except Exception as e:
    print(f"Error in voice verification: {e}")
```

```python
print("\n=== Testing finding similar voices ===")
try:
    print(f"Finding voices similar to {voice1_path}...")
    matches = DeepVoice.find(voice1_path, [voice2_path])

    print(f"Found {len(matches)} matches.")

    if matches:
        print("\nFirst match details:")
        match = matches[0]
        print(f"  - Voice: {match['voice'] if 'voice' in match else 'N/A'}")
        print(f"  - Confidence: {match['confidence'] if 'confidence' in match else 'N/A'}")
except Exception as e:
    print(f"Error in finding similar voices: {e}")
```
